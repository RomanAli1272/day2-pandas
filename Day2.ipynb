## Use apply() to transform a column based on a custom function.

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

# Define a custom function to calculate the sum of a row
def row_sum(row):
    return row.sum()

# Apply the custom function row-wise
df['Row_Sum'] = df.apply(row_sum, axis=1)

# Output the result
print(df)
```

## Use map() to replace values in a column based on a dictionary.

```python
df = pd.DataFrame({'col2': {0: 'a', 1: 2, 2: 0}, 'col1': {0: 'w', 1: 1, 2: 2}})
di = {1: "A", 2: "B"}
df['col1'].map(di) 
```
